First: Lets take a small sample of each csv file to see what we are dealing, since the Kaggle competition page does not give us any information about the data 

```python
import pandas as pd
import random
from IPython.display import display

sample_size = 0.05
random.seed = 1234 

def Sample_CSV(path, sample_size = .05):
    filename = path 
    n = sum(1 for line in open(filename)) - 1 #number of records in file (excludes header)
    sample = int(n * sample_size)
    print "# of Samples: ", sample
    skip = sorted(random.sample(xrange(1,n+1),n-sample)) #the 0-indexed header will not be included in the skip list
    df = pd.read_csv(filename, skiprows=skip)
    return df

train_numerical = Sample_CSV("../train_numeric.csv",sample_size)
display(train_numerical.head())

train_categorical = Sample_CSV("../train_categorical.csv",sample_size)
display(train_categorical.head())

train_date = Sample_CSV("../train_date.csv",sample_size)
display(train_date.head())
```

## Data Transformation 

Since the Dataset is so sparce I want to reorder the dataset. 

There is no point of the Same ID to be evaluated in more than one assembly line. I'm gonna rename all the Columns to Station_Feature and add one column that puts the assembly line in order to avoid losing information

```python
train_numerical.dtypes
```

```python
columns_names = train_numerical.drop(['Id','Response'],1).columns.values.tolist()
```

```python
test = 'L0_S0_F0'
print '_'.join((test.split('_')[-2:]))
```

```python
def ReShapeBosch(df):

    train_df = pd.DataFrame(columns=['Id','AssemblyLine','Response'])
    for Lines in range(0,4):
        # For each Line take the columns of that line and rename them 
        #tmp_df = train_numerical[columns_names]
        assembly_line = 'L'+ str(Lines)    
        matching = [column for column in columns_names if assembly_line in column]
        new_names = map(lambda x: '_'.join((x.split('_')[-2:])), matching)

        tmp_df = df[matching]
        tmp_df.columns = new_names

        tmp_df['AssemblyLine'] = Lines
        tmp_df['Response'] = df.Response
        tmp_df['Id'] = df.Id

        train_df = train_df.merge(tmp_df, how='outer', on=['Id','AssemblyLine','Response'])


    print "New DataFrame Shape", train_df.shape, "Old DataFrame Shape", train_numerical.shape
    
    return train_df
```

```python
train_df = ReShapeBosch(train_numerical)
display(train_df.head())
```
